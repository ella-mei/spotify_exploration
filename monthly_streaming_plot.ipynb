```python
import numpy as np
import matplotlib.pyplot as plt


# creating the dataset
data = {'January':1947, 'February':875, 'March':792,
        'April':990, 'May':902, 'June':1005, 'July':1020,
        'August':604, 'September':771, 'October':1815, 'November':2658,
        'December':1371,}
months = list(data.keys())
num_songs = list(data.values())

fig, ax = plt.subplots(figsize = (8, 5))

# creating the bar plot
plt.bar(months, num_songs, color ='plum',
        width = 0.4)


# Add labels and a title.
ax.set_xlabel('Month', labelpad=15, color='#333333')
ax.set_ylabel('No. Songs', labelpad=15, color='#333333')
ax.set_title('Number of Songs Streamed on Spotify per Month', pad=15, color='#333333',
             weight='bold')
plt.xticks(rotation=90)

# First, let's remove the top, right and left spines (figure borders)
# which really aren't necessary for a bar chart.
# Also, make the bottom spine gray instead of black.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')

# Second, remove the ticks as well.
ax.tick_params(bottom=False, left=False)

# Third, add a horizontal grid (but keep the vertical grid hidden).
# Color the lines a light gray as well.
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

fig.tight_layout()

plt.axhline(y=np.nanmean(num_songs), color='lightskyblue', label='Mean')

# creating the fred again dataset
data2 = {'January':107, 'February':17, 'March':15,
        'April':103, 'May':58, 'June':138, 'July':78,
        'August':31, 'September':74, 'October':225, 'November':260,
        'December':58,}
months = list(data.keys())
num_fred = list(data2.values())

plt.plot(months, num_fred, color='darkblue', label='Fred Again..')
plt.legend()
plt.show()



```
