```python
spotify_df2 = pd.read_csv(io.BytesIO(uploaded['StreamingHistory1.csv']))
```

```python
frames = [spotify_df, spotify_df2] # join both of the csv files together to get the full streaming history

spotify = pd.concat(frames)
```

```python
spotify.shape # check the DF shape, expecting (14750, 4) given the data
```

```python
# Begin to play with the data, adding an extra column to convert milliseconds to minutes in terms of play time. Then print the table head.

spotify['minsPlayed'] = spotify['msPlayed'].apply(lambda x: x*1/60000) # adding a minutes played column, by dividing by 60,000
spotify.head()
```

```python
df = spotify.drop(['msPlayed'], axis=1) # removing milliseconds played column as unneccesary
df
```

```python
# QUESTION 1: HOW MANY TOTAL MINUTES WERE PLAYED IN THE LAST YEAR FROM WHEN DATA EXTRACTED?|
```

```python
df['minsPlayed']
```

```python
sum_mins = sum(df['minsPlayed'])
sum_hours = sum_mins/60
sum_days = sum_hours/24
sum_weeks = sum_days/7

print(sum_weeks)
```

```python
# ANSWER 1: 745 HOURS, 9 MINUTES (TO THE NEAREST MINUTE) = 31 days, 4.435 weeks = basically 1 month of constant music!
```
