```python
frames = [spotify_df, spotify_df2] # join both csv files together to get full streaming history

spotify = pd.concat(frames)
```

```python
spotify.shape
```

```python
spotify['minsPlayed'] = spotify['msPlayed'].apply(lambda x: x*1/60000) # adding a minutes played column, by dividing by 60,000
spotify.head()
```

```python
df = spotify.drop(['msPlayed'], axis=1) # removing milliseconds played column
df
```

```python
sum_mins = sum(df['minsPlayed'])
sum_hours = sum_mins/60
sum_days = sum_hours/24
sum_weeks = sum_days/7

print(sum_weeks)
```
